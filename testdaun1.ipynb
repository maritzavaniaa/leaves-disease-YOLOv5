```python
# Path hasil deteksi
result_folder = "/content/drive/MyDrive/leaves_recognition_test/test_results1"
result_images = sorted(os.listdir(result_folder))

# Tampilkan beberapa hasil
for img_name in result_images[:5]:  # tampilkan 5 gambar pertama
    img_path = os.path.join(result_folder, img_name)
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(8,8))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(img_name)
    plt.show()
```
